```python
#import neccessary packages
from bs4 import BeautifulSoup
import requests
import time
import pandas as pd

find_stat = ['90', '1171', '546', '1222', '1204', '92', '422', '91', '30', '1192', '975', '977', '31', '569', '575', '33','32'
]
url_template = 'https://www.ncaa.com/stats/soccer-men/d1/current/team/{}/p{}'
time.sleep(5)

def get_table_data(team_number, page_number):
    try:
        url = url_template.format(team_number, page_number)
        response = requests.get(url, timeout=10)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')
        table = soup.find('table')
        team_data = pd.read_html(str(table))[0]
        return team_data
    except Exception as e:
        print(f"Error retrieving data for {team_number} - Page {page_number}: {e}")
        return None

# List to store data from all teams and pages
all_data = []

# Scrape data for each team and each page and append to the list
for team_number in find_stat:
    for page_number in range(1, 6):  # Adjust the range as needed
        team_data = get_table_data(team_number, page_number)
        if team_data is not None:
            all_data.append(team_data)

# Concatenate the list of DataFrames into a single DataFrame
result_table = pd.concat(all_data, ignore_index=True)

# Display the merged DataFrame
print(result_table)
```

```python
pd.DataFrame(result_table)
```

```python
#Fills the null values from the multiple team rows with appropiate values
result_table['Team'] = result_table['Team'].fillna(method='ffill')

# Convert non-numeric columns to numeric
result_table[['Rank', 'Assists', 'Team Games', 'Per Game', 'Team Min', 'GA', 'GAA', 'Goals', 'SoG', 'ShAtt', 'Pct.', 'Won', 'Loss', 'Tied']] = \
    result_table[['Rank', 'Assists', 'Team Games', 'Per Game', 'Team Min', 'GA', 'GAA', 'Goals', 'SoG', 'ShAtt', 'Pct.', 'Won', 'Loss', 'Tied']].apply(pd.to_numeric, errors='coerce')

# Group by the 'Team' column and aggregate using sum
merged_table = result_table.groupby('Team').sum().reset_index()

# Display the merged DataFrame
print(merged_table)
```

```python
df = pd.DataFrame(merged_table)
```
