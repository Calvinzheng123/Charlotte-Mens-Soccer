```python
sns.set(style = "whitegrid")

sorted_teams = combined_scored_df.groupby('TeamName')['xG'].median().sort_values(ascending=False).index

sorted_colors = [team_colors[team] for team in sorted_teams]

plt.figure(figsize=(13,8))
sns.boxplot(data = combined_scored_df, y = "TeamName", x = "xG", order=sorted_teams, palette=sorted_colors,medianprops={'color': 'white'})
plt.xticks(rotation=45)
plt.xlabel('Expected Goals (xG)')
plt.ylabel('Team')
plt.title('Distribution of Expected Goals (xG) for Each Goal Recorded by Team', fontsize=16, fontweight='bold')
plt.figtext(.01,.05, 'data from fbref', horizontalalignment='left', fontsize=10, alpha = .2)
plt.figtext(0.75, 0.5, 'Made by @CalvinZheng10', horizontalalignment='left', fontsize=9, fontweight = 'bold', alpha = .2)
# Show the plot
plt.show()
```
