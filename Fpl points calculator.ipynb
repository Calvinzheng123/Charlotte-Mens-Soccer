```python
# Define a dictionary to store points for each stat taken into account by fpl
points = {
    'goals': 0,
    'clean_sheet': 0,
    'saves': 0,
    'penalty_saves': 0,
    'assists': 0,
    'penalty_miss': 0,
    'goals_conceded': 0,
    'cards': 0,
    'minutes': 0,
    'own_goals': 0
}

# enter the number of minutes played
minutes_played = int(input("Minutes played: "))
if minutes_played <= 60:
    points['minutes'] = 1
else:
    points['minutes'] = 2

# enter player's position
position = input("Position (midfielder, striker, defender, goalie): ")

# Handle points calculation based on player's position
if position.lower() == 'striker':
    points['goals'] = int(input("Goals scored: ")) * 4
    points['assists'] = int(input("Assists: ")) * 3
    penalties_missed = input("Penalties missed? (yes/no): ")
    if penalties_missed.lower() == 'yes':
        points['penalty_miss'] = int(input("Penalties missed: ")) * -2
elif position.lower() == 'midfielder':
    points['goals'] = int(input("Goals scored: ")) * 5
    points['assists'] = int(input("Assists: ")) * 3
    clean_sheet = input("Clean sheet? (yes/no): ")
    if clean_sheet.lower() == 'yes':
        points['clean_sheet'] = 1
elif position.lower() == 'defender':
    points['goals'] = int(input("Goals scored: ")) * 4
    points['assists'] = int(input("Assists: ")) * 3
    clean_sheet = input("Clean sheet? (yes/no): ")
    if clean_sheet.lower() == 'yes':
        points['clean_sheet'] = 4
    goals_conceded = int(input("Goals conceded: "))
    points['goals_conceded'] = -(goals_conceded // 2)
elif position.lower() == 'goalie':
    points['goals'] = int(input("Goals scored: ")) * 4
    points['assists'] = int(input("Assists: ")) * 3
    clean_sheet = input("Clean sheet? (yes/no): ")
    if clean_sheet.lower() == 'yes':
        points['clean_sheet'] = 4
    points['saves'] = int(input("Saves: ")) // 3
    penalty_saves = input("Penalty saves? (yes/no): ")
    if penalty_saves.lower() == 'yes':
        points['penalty_saves'] = int(input("Penalty saves: ")) * 5
    goals_conceded = int(input("Goals conceded: "))
    points['goals_conceded'] = -(goals_conceded // 2)

# Calculate total points
total_points = sum(points.values())

# Print the total points
print("Total points:", total_points)
```
