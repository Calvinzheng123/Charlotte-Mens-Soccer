```python
#combine the dfs
scored_dfs = [
    Arsenal_Scored,
    AstonVilla_Scored,
    Bournemouth_Scored,
    Brentford_Scored,
    Brighton_Scored,
    Burnley_Scored,
    Chelsea_Scored,
    CrystalPalace_Scored,
    Everton_Scored,
    Fulham_Scored,
    Liverpool_Scored,
    LutonTown_Scored,
    ManCity_Scored,
    ManUnited_Scored,
    Newcastle_Scored,
    NottinghamForest_Scored,
    SheffieldUnited_Scored,
    Spurs_Scored,
    WestHam_Scored,
    Wolves_Scored
]

combined_scored_df = pd.concat(scored_dfs)
```

```python
#turning the data in team_goal_counts to show the scoring and conceding each interval
import pandas as pd

time_intervals = ["0-15", "16-30", "31-45", "46-60", "61-75", "76-90", "90+"]
team_goal_counts = pd.DataFrame(columns=time_intervals)


# Function to convert '90+' to '90' and '45+' to '45' before converting to integer
def convert_minute_to_int(minute_str):
    if '+' in minute_str:
        return int(minute_str.replace('+', ''))
    else:
        return int(minute_str)

# Apply the function to convert 'Minute' column to integers
combined_scored_df['Minute'] = combined_scored_df['Minute'].apply(convert_minute_to_int)

for team in combined_scored_df['TeamName'].unique():
    team_data = combined_scored_df[combined_scored_df['TeamName'] == team]
    
    goals_by_time_interval = []
    for interval in time_intervals:
        if interval == '90+':  # Handle the special case
            goals_count = len(team_data[team_data['Minute'] >= 90])
        else:
            start, end = map(int, interval.split('-'))
            goals_count = len(team_data[(team_data['Minute'] >= start) & (team_data['Minute'] <= end)])
        goals_by_time_interval.append(goals_count)
    
    team_goal_counts.loc[team] = goals_by_time_interval
```

```python
#your dataframe wont have a team column so you need to add it
team_names = team_goal_counts.index.tolist()  
team_goal_counts.index = team_names  
team_goal_counts.insert(0, "Team", team_names) 
```
