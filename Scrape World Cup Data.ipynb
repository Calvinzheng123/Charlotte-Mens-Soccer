```python
import requests
import pandas as pd

# List of match IDs
match_ids = [
    10230580, 10230635, 10230637, 10230581, 10230634, 10230640, 10230639, 
    10230631, 10230636, 10230579, 10230632, 10230633, 10230582, 10230641, 
    10230638, 10230551, 10230538, 10230537, 10230529, 10230541, 10230533, 
    10230532, 10230536, 10230556, 10230557, 10230543, 10230548, 10230547, 
    10230546, 10230542, 10230540, 10230528, 10230565, 10230555, 10230560, 
    10230574, 10230577, 10230576, 10230563, 10230554, 10230573, 10230564, 
    10230575, 10230562, 10230571, 10230535, 10230531, 10230545, 10230550, 
    10230570, 10230559, 10230553, 10230558, 10230568, 10230561, 10230569, 
    10230567, 10230544, 10230549, 10230552, 10230539, 10230530, 10230534
]

# Initialize an empty list to store DataFrames
dfs = []

# Loop through match IDs
for match_id in match_ids:
    # Make request with current match ID
    response = requests.get(f'https://api.sofascore.com/api/v1/event/{match_id}/shotmap')
    
    # Check if request was successful
    if response.status_code == 200:
        # Parse JSON response
        shots = response.json()
        
        # Check if shotmap data exists
        if shots.get('shotmap'):
            # Convert shotmap data to a DataFrame and append to list
            df = pd.json_normalize(shots['shotmap'])
            dfs.append(df)
        else:
            print(f"No shotmap found for match ID: {match_id}")
    else:
        print(f"Failed to fetch data for match ID: {match_id}")

# Concatenate all DataFrames into a single DataFrame
combined_df = pd.concat(dfs, ignore_index=True)

# Display the combined DataFrame
print(combined_df)
```
